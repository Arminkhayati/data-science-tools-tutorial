```python
import pandas as pd
import numpy as np
from numpy import nan as NA
df = pd.DataFrame(np.random.random((7,3)))
df
df.iloc[:4, 1] = NA
df.iloc[:2, 2] = NA
df
```

```python
df.fillna(0)
```

```python
# Calling fillna with a dict, you can use a different fill value for each column

# df = df.fillna({1: 0.5, 2: 1000})
# or :
df.fillna({1: 0.5, 2:1000}, inplace=True)
df
```

```python
df = pd.DataFrame(np.random.random((6,3)))
df.iloc[2: , 1] = NA
df.iloc[4: , 2] = NA
df
```

```python
df.fillna(method='ffill')
```

```python
df.fillna(method='ffill', limit=2)
```

```python
data = pd.Series([1., NA, 3.5, NA, 7])
data
```

```python
data.fillna(data.mean())
```
