```python
import pandas as pd
import numpy as np
from numpy import nan as NA


data = pd.DataFrame({'food': ['bacon', 'pulled pork', 'bacon',
                              'Pastrami', 'corned beef', 'Bacon','pastrami',
                              'honey ham', 'nova lox'],
                              'ounces': [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})
data
```

```python
# add a column indicating the type of animal that each food
# came from. Let’s write down a mapping of each distinct meat type to the kind of
# animal:
meat_to_animal = {
'bacon': 'pig',
'pulled pork': 'pig',
'pastrami': 'cow',
'corned beef': 'cow',
'honey ham': 'pig',
'nova lox': 'salmon'
}
meat_to_animal
```

```python
lower_cased = data['food'].str.lower()
lower_cased
```

```python
#  The map method on a Series accepts a function or dict-like object containing a map‐
# ping
data['animal'] = lower_cased.map(meat_to_animal)
data
```

```python
# or :
data['animal'] = data['food'].map(lambda x: meat_to_animal[x.lower()])
data
```

```python
def to_map(x):
    return meat_to_animal[x.lower()]
data
```

```python
data['animal'] = data['food'].map(to_map)
data
```
