```python
import pandas as pd
import numpy as np


data = pd.read_csv('/home/armin/Data/Development/Python/pydata-book-2nd-edition/examples/macrodata.csv')
data.head()
```

```python
periods = pd.PeriodIndex(year=data.year, quarter=data.quarter,name='date')
columns = pd.Index(['realgdp', 'infl', 'unemp'], name='item')
data = data.reindex(columns=columns)
data.index = periods.to_timestamp('D', 'end')
data[:5]
```

```python
data.stack()[:15]
```

```python
'''

This is useful when the index needs to be treated as a column, or when the index is 
meaningless and needs to be reset to the default before another operation

'''

data.stack().reset_index()[:5]
```

```python
ldata = data.stack().reset_index().rename(columns={0: 'value'})
ldata[:10]
```

```python
'''
you might prefer to have a DataFrame containing
one column per distinct item value indexed by timestamps in the date column. Data‐
Frame’s pivot method performs exactly this transformation
The first two values passed are the columns to be used respectively as the row and
column index, then finally an optional value column to fill the DataFrame.
'''

pivoted = ldata.pivot('date', 'item', 'value')
pivoted[:10]
```

```python
'''
Suppose
you had two value columns that you wanted to reshape simultaneously
'''
ldata['value2'] = np.random.randn(len(ldata))
ldata[:5]
```

```python
pivoted = ldata.pivot('date', 'item')
pivoted[:5]
```

```python
pivoted['value'][:5]
```

```python
pivoted = ldata.pivot('date', 'item','value2')
pivoted[:5]
```

```python
'''Note that pivot is equivalent to creating a hierarchical index using set_index fol‐
lowed by a call to unstack
'''
unstacked = ldata.set_index(['date', 'item'])
unstacked[:7]
```

```python
unstacked = ldata.set_index(['date', 'item']).unstack('item')
unstacked[:7]
```
